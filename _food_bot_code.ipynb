```python
import pandas as pd
import torch
from transformers import BertTokenizer, BertForSequenceClassification
import telebot

# Load the dataset
data = pd.read_excel('Disadvantages_of_Eating_Fast_Foods.xlsx')

# Prepare the data for training
labels = data['Fast Food Item']
disadvantages = data['Disadvantages']
effects = data['Effects']

# Load the pre-trained BERT model and tokenizer
model_name = 'bert-base-uncased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name, num_labels=len(labels.unique()))

# Set up the Telegram bot
bot = telebot.TeleBot("6118195764:AAExI0zPLqnSyffbp6U7J6sumCgIz9qPuIk")

# Define the Telegram bot command
@bot.message_handler(commands=['start'])
def start_command(message):
    bot.reply_to(message, "Welcome to the Fast Food Bot! Please enter a fast food item.")

# Handle user messages
@bot.message_handler(func=lambda message: True)
def handle_message(message):
    fast_food_item = message.text

    # Find the index of the fast food item in the dataset
    index = labels[labels.str.lower() == fast_food_item.lower()].index

    if not index.empty:
        # Get the disadvantages and effects for the specific fast food item
        disadvantages_text = disadvantages[index[0]]
        effects_text = effects[index[0]]

        # Reply with the disadvantages and effects
        reply_message = f"Disadvantages of {fast_food_item}:\n\n{disadvantages_text}\n\nEffects:\n{effects_text}"
    else:
        reply_message = f"No information found for {fast_food_item}."

    # Send the reply message
    bot.reply_to(message, reply_message)

# Start the Telegram bot
bot.polling()
```
